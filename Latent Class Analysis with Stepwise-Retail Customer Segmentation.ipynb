```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Number of customers
n_customers = 1000

# Purchase Frequency
purchase_frequency = np.random.choice(['Daily', 'Weekly', 'Monthly'], size=n_customers, p=[0.1, 0.4, 0.5])

# Average Purchase Amount
purchase_amount = np.random.choice(['Low', 'Medium', 'High'], size=n_customers, p=[0.5, 0.3, 0.2])

# Product Categories
product_categories = np.random.choice(['Electronics', 'Clothing', 'Groceries'], size=n_customers, p=[0.3, 0.4, 0.3])

# Payment Method
payment_method = np.random.choice(['Cash', 'Credit Card', 'Digital Wallet'], size=n_customers, p=[0.4, 0.4, 0.2])

# Customer Loyalty
customer_loyalty = np.random.choice(['Member', 'Non-Member'], size=n_customers, p=[0.3, 0.7])

# Create DataFrame
data = pd.DataFrame({
    'Purchase Frequency': purchase_frequency,
    'Average Purchase Amount': purchase_amount,
    'Product Categories': product_categories,
    'Payment Method': payment_method,
    'Customer Loyalty': customer_loyalty
})

data.head()
```

```python
# Save to CSV
data.to_csv('synthetic_customer_data.csv', index=False)
```

```python
# Install the stepmix library
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from stepmix.stepmix import StepMix

# Load the synthetic dataset
data = pd.read_csv('synthetic_customer_data.csv')

# Encode categorical variables
le = LabelEncoder()
data_encoded = data.apply(le.fit_transform)

# Convert to numpy array
X = data_encoded.values

# Fit StepMix model
model = StepMix(n_components=3, n_steps=1)
model.fit(X)

# Predict latent classes
latent_classes = model.predict(X)
data['Latent Class'] = latent_classes

data.head()
```

```python
# Analyze the distribution of latent classes
class_distribution = data['Latent Class'].value_counts()
print(class_distribution)

# Group by latent class to see characteristics of each class
class_characteristics = data.groupby('Latent Class').mean()
print(class_characteristics)
```
